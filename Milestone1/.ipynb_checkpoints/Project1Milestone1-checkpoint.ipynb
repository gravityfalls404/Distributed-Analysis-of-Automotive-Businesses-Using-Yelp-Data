SQL Queries Start from here



```python
# Fetch basic details of top-rated Health & Medical businesses
spark.sql("""     
SELECT business_id, name, stars, review_count
FROM health
ORDER BY stars DESC
LIMIT 10;
          """).show()
```

```python
import matplotlib.pyplot as plt

# Run query and convert result to pandas DataFrame
top_rated_df = spark.sql("""
    SELECT business_id, name, stars, review_count
    FROM health
    ORDER BY stars DESC
    LIMIT 10
""").toPandas()

# Plot
plt.figure(figsize=(10, 6))
plt.barh(top_rated_df['name'], top_rated_df['stars'], color='skyblue')
plt.xlabel("Stars")
plt.ylabel("Business Name")
plt.title("Top 10 Health & Medical Businesses by Rating")
plt.gca().invert_yaxis()
plt.show()
```

```python
spark.sql("""     
SELECT city, COUNT(business_id) AS business_count
FROM health
GROUP BY city
ORDER BY business_count DESC;

          """).show()
```

```python
# Run query and convert result to pandas DataFrame
business_count_city_df = spark.sql("""
    SELECT city, COUNT(business_id) AS business_count
    FROM health
    GROUP BY city
    ORDER BY business_count DESC
""").toPandas()

# Plot
plt.figure(figsize=(10, 8))
plt.barh(business_count_city_df['city'], business_count_city_df['business_count'], color='lightgreen')
plt.xlabel("Business Count")
plt.ylabel("City")
plt.title("Count of Health & Medical Businesses by City")
plt.gca().invert_yaxis()
plt.show()
```

```python
spark.sql("""     
SELECT b.postal_code, AVG(b.stars) AS avg_rating, COUNT(b.business_id) AS total_businesses, COUNT(c.date) AS total_checkins
FROM health b
LEFT JOIN checkin c ON b.business_id = c.business_id
GROUP BY b.postal_code
ORDER BY avg_rating DESC;

          """).show()
```

```python
# Run query and convert result to pandas DataFrame
avg_rating_zip_df = spark.sql("""
    SELECT b.postal_code, AVG(b.stars) AS avg_rating, COUNT(b.business_id) AS total_businesses, COUNT(c.date) AS total_checkins
    FROM health b
    LEFT JOIN checkin c ON b.business_id = c.business_id
    GROUP BY b.postal_code
    ORDER BY avg_rating DESC
""").toPandas()

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(avg_rating_zip_df['avg_rating'], avg_rating_zip_df['total_checkins'], alpha=0.7, color='purple')
plt.xlabel("Average Rating")
plt.ylabel("Total Check-Ins")
plt.title("Average Rating vs. Total Check-Ins per ZIP Code")
plt.show()
```

```python
spark.sql("""     
    SELECT 
        CASE 
            WHEN b.attributes.WiFi IS NOT NULL THEN 'WiFi'
            WHEN b.attributes.GoodForKids IS NOT NULL THEN 'Good for Kids'
            WHEN b.attributes.WheelchairAccessible IS NOT NULL THEN 'Wheelchair Accessible'
            ELSE 'Other'
        END AS attribute_type,
        AVG(b.stars) AS avg_rating,
        COUNT(b.business_id) AS business_count
    FROM health b
    GROUP BY attribute_type
    ORDER BY avg_rating DESC

          """).show()
```

```python
# Run query and convert result to pandas DataFrame
attribute_specific_rating_df = spark.sql("""
    SELECT 
        CASE 
            WHEN b.attributes.WiFi IS NOT NULL THEN 'WiFi'
            WHEN b.attributes.GoodForKids IS NOT NULL THEN 'Good for Kids'
            WHEN b.attributes.WheelchairAccessible IS NOT NULL THEN 'Wheelchair Accessible'
            ELSE 'Other'
        END AS attribute_type,
        AVG(b.stars) AS avg_rating,
        COUNT(b.business_id) AS business_count
    FROM health b
    GROUP BY attribute_type
    ORDER BY avg_rating DESC
""").toPandas()

# Plot
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(attribute_specific_rating_df['attribute_type'], attribute_specific_rating_df['avg_rating'], color='lightblue')
plt.xlabel("Attribute Type")
plt.ylabel("Average Rating")
plt.title("Average Rating by Key Business Attributes in Health & Medical Category")
plt.xticks(rotation=45)
plt.show()

```

```python
spark.sql("""     
SELECT r.review_id, r.user_id, r.useful, r.text, b.name AS business_name, b.stars AS business_rating
FROM review r
JOIN health b ON r.business_id = b.business_id
WHERE b.stars >= 4
ORDER BY r.useful DESC
LIMIT 20;

          """).show()
```

```python
# Run query and convert result to pandas DataFrame
useful_reviews_df = spark.sql("""
    SELECT r.review_id, r.user_id, r.useful, r.text, b.name AS business_name, b.stars AS business_rating
    FROM review r
    JOIN health b ON r.business_id = b.business_id
    WHERE b.stars >= 4
    ORDER BY r.useful DESC
    LIMIT 20
""").toPandas()

# Plot
plt.figure(figsize=(10, 6))
plt.barh(useful_reviews_df['business_name'], useful_reviews_df['useful'], color='salmon')
plt.xlabel("Useful Score")
plt.ylabel("Business Name")
plt.title("Top 20 Most Useful Reviews for Top-Rated Health & Medical Businesses")
plt.gca().invert_yaxis()
plt.show()
```
